```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sp
from scipy.stats import norm

x=np.arange(-3,3,0.01)
plt.plot(x,norm.pdf(x))
plt.show()
```

```python
plt.plot(x,norm.pdf(x))
plt.plot(x,norm.pdf(x,1,0.5))
plt.show()
```
